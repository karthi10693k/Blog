```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
population=np.random.uniform(1,5000,1000)#Population is uniform distribution of 1000 recods  
```

```python
sns.distplot(population)
```

```python
np.mean(population)
```

```python
np.std(population)
```

**sample size should be greater than or equal to 30 as per the Central Limit Theorem**

```python
sample = []
for i in range(0,30):
    sample.append(random.choice(population))
```

```python
len(sample)
```

```python
np.mean(sample)  #sample mean of single sample 
```

**We need to take more sample to form the sampling distribution of mean**

**Sample size :30**

```python
Sample_distribution_mean=[]
for i in range(0,50):
    sample = []
    for i in range(0,30):
        sample.append(random.choice(population))
    Sample_distribution_mean.append(np.mean(sample))
```

```python
np.mean(Sample_distribution_mean)
```

```python
np.std(Sample_distribution_mean)
```

```python
sns.distplot(Sample_distribution_mean)
```
