```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
```

Lets create a population uniformly distriubuted with total of 10000 quantity 

```python
population=np.random.uniform(1,5000,10000)#Population is uniform distribution of 10000 recods  
```

```python
sns.distplot(population)
```

The Mean of Population : **2477.49** 
<br>
The Standard Deviation : **1422.16** 

```python
np.mean(population)
```

```python
np.std(population)
```

**Lets take 30 sample from population and name it as single sample_set** <br>
Random sample is take from population using **random.choice()** function <br>
<br>
**sample _set** is sample made of __30 samples__ 

```python
sample_set = []
for i in range(0,30):
    sample_set.append(random.choice(population))
```

```python
len(sample_set)
```

```python
np.mean(sample_set)  #sample mean of single sample_set
```

```python
sns.distplot(sample_set)
```

**We need to take more sample set  to form the sampling distribution of mean**<br>
there is no thu,b

**Sample size :30**

```python
Sample_distribution_mean=[]
for i in range(0,1000):
    sample_set = []
    for i in range(0,30):
        sample_set.append(random.choice(population))
    Sample_distribution_mean.append(np.mean(sample_set))
```

```python
sns.distplot(Sample_distribution_mean)
```

```python
np.mean(Sample_distribution_mean)
```

```python
np.std(Sample_distribution_mean)
```

Plot the frequency distribution of the means of the 1000 sample ,<br>We will get normal distribution curve and the number of sample increase the distribution trend to become perfect normal distribution,<br>
The mean approach close to the population mean as the sample size increase  

```python
sns.distplot(Sample_distribution_mean)
```
